## Part1: CSNN

```python
train_acc_path = "results/CSNN-train-acc.csv"
train_loss_path = "results/CSNN-train-loss.csv"
val_acc_path = "results/CSNN-val-acc.csv"
val_loss_path = "results/CSNN-val-loss.csv"
```

```python
import pandas as pd

train_acc = pd.read_csv(train_acc_path)
train_loss = pd.read_csv(train_loss_path)
val_acc = pd.read_csv(val_acc_path)
val_loss = pd.read_csv(val_loss_path)
```

```python
train_acc
```

```python
# remove MIN and MAX columns
import re

train_acc = train_acc.drop(columns=[col for col in train_acc.columns if re.search(r'.*(MIN|MAX)', col)])
train_loss = train_loss.drop(columns=[col for col in train_loss.columns if re.search(r'.*(MIN|MAX)', col)])
val_acc = val_acc.drop(columns=[col for col in val_acc.columns if re.search(r'.*(MIN|MAX)', col)])
val_loss = val_loss.drop(columns=[col for col in val_loss.columns if re.search(r'.*(MIN|MAX)', col)])
```

```python
train_acc
```

```python
# clean column names

train_acc.columns = ['iteration', 'ResNet-18', 'VGG-19', 'VGG-8', 'VGG-16']
train_loss.columns = ['iteration', 'ResNet-18', 'VGG-19', 'VGG-8', 'VGG-16']
val_acc.columns = ['iteration', 'ResNet-18', 'VGG-19', 'VGG-8', 'VGG-16']
val_loss.columns = ['iteration', 'ResNet-18', 'VGG-19', 'VGG-8', 'VGG-16']
```

```python
train_acc
```

```python
# permute columns: VGG-8, VGG-16, VGG-19, ResNet-18
train_acc = train_acc[['iteration', 'VGG-8', 'VGG-16', 'VGG-19', 'ResNet-18']]
train_loss = train_loss[['iteration', 'VGG-8', 'VGG-16', 'VGG-19', 'ResNet-18']]
val_acc = val_acc[['iteration', 'VGG-8', 'VGG-16', 'VGG-19', 'ResNet-18']]
val_loss = val_loss[['iteration', 'VGG-8', 'VGG-16', 'VGG-19', 'ResNet-18']]
train_acc
```

```python
# plot curves
# 4 figures: train acc, train loss, val acc, val loss

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)

for col in train_acc.columns[1:]:
    plt.plot(train_acc['iteration'], train_acc[col], label=col, linewidth=0.8)
plt.title('Train Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.legend()
plt.subplot(2, 2, 2)

for col in train_loss.columns[1:]:
    plt.plot(train_loss['iteration'], train_loss[col], label=col, linewidth=0.8)
plt.title('Train Loss')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()
plt.subplot(2, 2, 3)

for col in val_acc.columns[1:]:
    plt.plot(val_acc['iteration'], val_acc[col], label=col, linewidth=0.8)
plt.title('Validation Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.legend()
plt.subplot(2, 2, 4)

for col in val_loss.columns[1:]:
    plt.plot(val_loss['iteration'], val_loss[col], label=col, linewidth=0.8)
plt.title('Validation Loss')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('results/csnn-training-curves.pdf', dpi=300)
```

```python
# compute best validation acc
print(val_acc['VGG-8'].max())
print(val_acc['VGG-16'].max())
print(val_acc['VGG-19'].max())
print(val_acc['ResNet-18'].max())
```

## Part2: surrogate gradients

```python
# plot surrogate function with different alphas (2, 4, 6, 8, 10)
# sigmoid: h(x) = alpha * sigmoid(x) * (1-sigmoid(x))

# plot
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 1000)
alphas = [2, 4, 6, 8, 10]
colors = ["#9ed3f8", "#7faef3", "#f69a9a", "#e85656", "#961423"]

plt.figure(figsize=(10, 6))
for alpha, color in zip(alphas, colors):
    sigmoid = 1 / (1 + np.exp(-x))
    h = alpha * sigmoid * (1 - sigmoid)
    plt.plot(x, h, label=f'alpha={alpha}', color=color)
plt.xlabel('$x$')
plt.ylabel('$h(x)$')
plt.legend()
plt.grid()
plt.savefig('results/surrogate-sigmoid.pdf', dpi=300)

# superspike: h(x) = 1 / (1 + alpha * abs(x))^2
plt.figure(figsize=(10, 6))
for alpha, color in zip(alphas, colors):
    h = 1 / (1 + alpha * np.abs(x))**2
    plt.plot(x, h, label=f'alpha={alpha}', color=color)
plt.xlabel('$x$')
plt.ylabel('$h(x)$')
plt.legend()
plt.grid()
plt.savefig('results/surrogate-superspike.pdf', dpi=300)

# triangular: h(x) = max(0, 1 - alpha * abs(x))
plt.figure(figsize=(10, 6))
for alpha, color in zip(alphas, colors):
    h = np.maximum(0, 1 - alpha * np.abs(x))
    plt.plot(x, h, label=f'alpha={alpha}', color=color)
plt.xlabel('$x$')
plt.ylabel('$h(x)$')
plt.legend()
plt.grid()
plt.savefig('results/surrogate-triangular.pdf', dpi=300)
```
